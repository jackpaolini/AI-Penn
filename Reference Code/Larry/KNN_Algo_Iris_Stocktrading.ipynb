## <font color=' #7cb342 '>Supervised Learning vs Unsupervised Learning</font>

Supervised learning is the most common.  Its observed data is associated with a label (either observed or specified by experts).  Once learned, the model will be used to predict label on new unseen data.
<br><br>


Unsupervised learning involves modeling the features of a dataset without reference to any label.  Hence it is often described as "letting the dataset speak for itself." These models include structurable tasks such as clustering (identifying distinct groups of data).



## <font color=' #7cb342 '>Supervised Learning: Classification vs. Regression</font>

When you are predicting a type, a category or a label, you are performing **Classification** task.
<br>
When you are predicting a numerical value or a continuous quantity, you are typically performing **Regression** task.
<br><br>
For example, predicting stock price is a **Regression** task, while predicting the winner of a presidential election is **Classification** task.

## <font color=' #7cb342 '>KNN (K-Nearest Neighbors) Model</font>

KNN algorithms <u>classify new data points based on a similarity measures (e.g. distance function)</u>.  **Classification is done by a majority vote to its neighbors**:  the label is assigned to t**he class which has the most nearest neighbors**.

![alt text](http://res.cloudinary.com/dyd911kmh/image/upload/f_auto,q_auto:best/v1531424125/KNN_final1_ibdm8a.png)

## KNN using Scikit-learn Machine Learning Library

Very popular machine learning library in Python.  Provide many machine learning models in easy-to-use ways.

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import  accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

#### Split data into features and labels 

```
features = df[['sepal length in cm', 
               'sepal width in cm', 
               'petal length in cm', 
               'petal width in cm']]
labels = df['class']
```

#### Convert class labels into numbers

Class labels are strings.  We need to convert them into numbers that could be understood by computers.

```
encoder = LabelEncoder()
y = encoder.fit_transform(labels)
```

As you can see Iris-setosa Iris-versicolor Iris-virginica are converted to 0, 1, 2 respectively.

```
y
```

#### Generate training set and test set

It is common practice when performing a (supervised) machine learning training to hold out part of the available data as a test set.  sk-learn will help us quickly sample a training set while holding out certain percentage of data for testing (evaluating) our model.  The command below will have a 7-3 split of training data vs. testing data.

```
# x_train, y_train will be used for training the model
# while x_test, y_test will be used for testing/evaluating the trained model
# random_state is the seed used by random number generator
# stratify=y ensures data is split in the stratified fashion
x_train, x_test, y_train, y_test = train_test_split(features, y, test_size = 0.3, 
                                                    random_state = 101, stratify=y)
```

The data count should show stratified effect.

```
print('Labels counts in y:', np.bincount(y))
print('Labels counts in y_train:', np.bincount(y_train))
print('Labels counts in y_test:', np.bincount(y_test))
```

#### Feature Scaling

Feature scaling is a method used to standardize the range of independent variables or features of data. In data processing, it is also known as data normalization and is generally performed during the data preprocessing step.

```
x_train.head()
```

```
sc = StandardScaler()

sc.fit(features)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)

type(x_train_std)
```

```
x_train_std[:5]
```

#### Create KNN model

For parameter p below.  When p = 1, this is equivalent to using manhattan_distance, and euclidean_distance for p = 2 (default).

```
# K=3 below
knn_model = KNeighborsClassifier(n_neighbors=7, p=2)
```

#### Run KNN Machine Learning Process (Train model, Fit data)

```
knn_model.fit(x_train_std, y_train);
```

#### Run Prediction on test set

```
knn_predict = knn_model.predict(x_test_std)
```

#### Is the prediction accurate?

## Visualize Decision/Plot Decision Regions

It is hard to visualize Machine Learning, which is often called as **Black Box**.  We will try to visualize the result below.

### Helper functions.  Just run and skip.

```
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')

    # highlight test samples
    if test_idx:
        # plot all samples
        X_test, y_test = X[test_idx, :], y[test_idx]
        c_test = [colors[i] for i in y_test]

        plt.scatter(X_test[:, 0],
                    X_test[:, 1],
                    c=c_test,
                    edgecolor='black',
                    alpha=0.8,
                    linewidth=1,
                    marker='o',
                    s=100, 
                    label='test set')
```

### Pick 2 features to visualize Decision Boundary

We will pick features: Petal Length and Petal Width.

```
X_train_std = x_train_std[:, 2:]
X_test_std  = x_test_std[:, 2:]

# Fit KNN model again
knn_model.fit(X_train_std, y_train);
```

Now visualized the Decision Boundary.

```
X_combined = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))

plot_decision_regions(X_combined, y_combined, 
                      classifier=knn_model, test_idx=range(105, 150))

plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
```

## KNN model: Pros and Cons

###Pros

1.   It is extremely easy to implement
2.   It is, so called, lazy learning algorithm and therefore requires no training prior to making real time predictions. This makes the KNN algorithm much faster than other algorithms that require training.
3.   Since the algorithm requires no training before making predictions, new data can be added seamlessly.
4. It's easy to understand.

### Cons



1.   The KNN algorithm doesn't work well with high dimensional data because with large number of dimensions, it becomes difficult for the algorithm to calculate distance in each dimension.
2.   The KNN algorithm has a high prediction cost for large datasets. This is because in large datasets the cost of calculating distance between new point and each existing point becomes higher.
3. Finally, the KNN algorithm doesn't work well with categorical features since it is difficult to find the distance between dimensions with categorical features.

We could create a Auto Stock Trading Strategy using KNN model.  Such a trading strategy is also known as, <b>Machine Trading Strategy</b>, <b>Algorithmic Trading Strategy</b> (or <b>Algo Trading Strategy</b>), <b>Black Box Trading Strategy</b> or <b>Quantitative Trading Strategy</b>.

As we all know, it's hard to create a consistent money maker.  Our strategy is not an exception either.  It's a basic, single strategy.  To be used in real life, it will need a lot of tuning, backtesting and most likely be combined with other trading strategies and risk managements.

But for now, let's go ahead to create our first Algo Trading Strategy!

## Business Requirements and Trading Strategy Specification

### Which Stock to Trade?

We will trade stock [SPY](https://finance.yahoo.com/quote/SPY?p=SPY&.tsrc=fin-srch).  This stock tracks [S&P 500 index](https://en.wikipedia.org/wiki/S%26P_500_Index).
<br><br>
Our strategy should be able to work on any stocks.  Noted that different stocks might need different model parameters

### Trading Strategy Specification

Our strategy will use prior 3 month stock data to build KNN model and trade stock daily for the next month.
<br><br>
At the end of the next month, we will then rebuild our KNN model using the latest 3 month stock data and use the updated model to trade the next next month.
<br><br>
We plan to continue trading like this forever.


### Initial Trading Strategy Development and Ongoing Maintenance

If the model is not working very well for a couple of month, we might tune our strategy by changing the parameter of our Machine Learning model.  Specifically, we might change the number of neighbors parameter of KNN model and back-test to make sure it works.
<br><br>
We might also change the time period for Machine Learning model input (currently it's 3 month) and/or the Trading time period (currently 1 month).

## Data Acquisition

We will use pandas_datareader module to load daily stock price from financial web site: http://finance.yahoo.com 

```
from pandas_datareader import data as pdr
```

Load daily stock price from 2018-09-01 to 2019-01-01.  The data from the first 3 month (2018-09-01 to 2018-12-01) will be used for model input.  The data for the last month (2018-12-01 to 2019-01-01) will be used for simulated trading.

```
df = pdr.get_data_yahoo('SPY', '2018-09-01', '2019-01-01')
```

## Data Cleaning/Preprocessing

```
df = df.dropna()
```

```
df = df[['Open', 'High', 'Low', 'Close']]
```

```
df.head()
```

## Feature Engineering

```
# With feature scaling
#df['Open-Close'] = 100*(df.Open - df.Close)/df.Close.shift(-1)
#df['High-Low'] = 100*(df.High - df.Low)/df.Close.shift(-1)

# Without feature scaling
df['Open-Close'] = df.Open - df.Close
df['High-Low'] = df.High - df.Low
df = df.dropna()
X=df[['Open-Close', 'High-Low']]
X.head()
```

## Trading Signal

The target variable is whether tomorrow's closing price is higher than today's closing price.
<br><br>
This target variable is our trading signal:
> If signal tells us that tomorrow's closing price will be **higher**, then the strategy will **buy** stock at today's closing price and **sell** stock at tomorrow's closing price.  
>
>If signal tells us that tomorrow's closing price will be **lower**, then the strategy will **sell** stock at today's closing price and **buy back** the stock to tomorrow's closing price.


```
Y = np.where(df['Close'].shift(-1)>df['Close'], 1, -1)
```

## Prepare Model Input and Simulated Trading Data

```
split_percentage = 0.75
split = int(split_percentage * len(df))

#With 4 month data, the first 75% data (3 month) will be for model input
X_train = X[:split]
Y_train = Y[:split]

#The last 25% data (1 month) will be for simulated trading/testing
X_test = X[split:]
Y_test = Y[split:]
```

## Train KNN Model

```
knn = KNeighborsClassifier(n_neighbors=25)

knn.fit(X_train, Y_train)

accuracy_train = accuracy_score(Y_train, knn.predict(X_train))

print('Train_data Accuracy: %.2f' %accuracy_train)
```

## Simulate Trading

```
accuracy_test = accuracy_score(Y_test, knn.predict(X_test))

print('Test_data Accuracy: %.2f' %accuracy_test)
```

Calculate cumulative return

```
df['Predicted_Signal'] = knn.predict(X)

df['SPY_Returns'] = np.log(df['Close']/df['Close'].shift(1))
Cumulative_SPY_Returns = df[split:]['SPY_Returns'].cumsum()*100

df['KNN_Strategy_Returns'] = df['SPY_Returns'] * df['Predicted_Signal'].shift(1)
Cumulative_KNN_Strategy_Returns = df[split:]['KNN_Strategy_Returns'].cumsum()*100

print("Simulate trading from "+str(X[split:].index[0])[:10]+" to "+str(X[split:].index[-1])[:10]+" :")
print('KNN Strategy Extra Return Percentage = %.2f' %(Cumulative_KNN_Strategy_Returns[-1]-Cumulative_SPY_Returns[-1]))

print("\n")
print(df[split:]['Predicted_Signal'])

plt.figure(figsize=(10,5))
plt.plot(Cumulative_SPY_Returns, color='r', label='SPY Returns')
plt.plot(Cumulative_KNN_Strategy_Returns, color='g', label='KNN Strategy Returns', alpha=0.5)
plt.legend()
plt.show()
```

## <font color='gold'>What could we take away from backtesting?</font>


1.   Quantitative Strategy could work in stock market
2.   Certain Strategy works well in certain periods, but not too well in other periods (for those periods, maybe other Strategies could do well)
3.   Even a simple Strategy, like the KNN strategy above, could work.  But stock market is different than Natural Science: it could change dramatically far more frequent.  Hence as a Algorithmic Trader, you need to keep monitoring your strategy and fine tune the parameters of your strategy.  Or, you might need to combine the strategy with other strategies. Or, be ready to abandon the strategy completely and switch to other new strategies.
