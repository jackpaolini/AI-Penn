*   Root Node
*   Recursive Partitioning

###  <font color=' #7cb342 '>Common Quantitative Partition Rules</font>

The decision rule, or the split criterion, is very important.  Common rules are Information Gain (Entropy) and Gini Index.

### <font color=' #7cb342 '>Recursive Partitioning</font>

Select the best attribute using Attribute Selection Measures(ASM) to split the records.

1.   Make that attribute a decision node and breaks the dataset into smaller subsets.
2.   Starts tree building by repeating this process recursively for each child until one of the condition will match:
            A.   All the tuples belong to the same attribute value.
            B.   There are no more remaining attributes.
            C.   There are no more instances.



## <font color=' red '>(Optional) Math Underlying</font>

###<font color=' #7cb342 '>(Optional) Information Gain</font>

Information Gain and Entropy are terminologies in Information Theory.  Decision Tree model relies on splitting to remove uncertainty of classification, and entropy is is a common measure of uncertainty.
<br><br>
Qualitatively, high entropy implies high uncertainty, while low entropy implies low uncertainty.
<br><br>
<img src="https://wikimedia.org/api/rest_v1/media/math/render/svg/bfe3616dee43f6287d4a4e2a557de8d48ad24926" width="400"/>
<br><br>
A 2-dimentional entroy graph looks like this:


<br><br>
<img src="http://lj-studios.com/big_data_club/decision_tree/entropy.png" width=500/>


Information gain is defined as the reduction of entropy (uncertainty) after the split.
<br><br>
Information gain (IG) is used as the criteria to select why feature to split and the splitting point for that feature for regression.
<br><br>
The recursive process goes:
<br>
1.  compute the entropy for data-set
2.  for every attribute/feature:
        A.  calculate entropy for all categorical values
        B.  take average information entropy for the current attribute
        C.  calculate gain for the current attribute
3.   pick the highest gain attribute.
4.   Repeat until we get the tree we desired.

#### Information Gain if splitting on Location

Entropy before splitting:
$$H(X) = -(p_{Mammal} * log_2(p_{Mammal}) + p_{Bird} * log_2(p_{Bird}))
              = -(\frac{4}{8}*log_2(\frac{4}{8}) + \frac{4}{8}*log_2(\frac{4}{8})) = 1 $$
<br>
After splitting on Location, we have <b>left node</b> having Location as Land:

| Color | Location | Class |
|----------|-------------|-----|
| Yellow | Tree | Bird |
| Red | Tree | Bird |
| Brown | Tree | Bird |

Entropy after splitting on Location:
<br><br>
$$\begin{eqnarray}
H(X|Land) &=& -(p_{Mammal|Land} &* log_2(p_{Mammal|Land}) &+ p_{Bird|Land} &* log_2(p_{Bird|Land})) \\
              &=& -(\frac{4}{5}&*log_2(\frac{4}{5}) &+ \frac{1}{5}&*log_2(\frac{1}{5})) \\
           &=& 0.721 \\
           \\
H(X|Tree) &=& -(p_{Mammal|Tree} &* log_2(p_{Mammal|Tree}) &+ p_{Bird|Tree} &* log_2(p_{Bird|Tree})) \\
              &=& -(\frac{3}{3}&*log_2(\frac{3}{3}) &+ \frac{0}{3}&*log_2(\frac{0}{3})) \\
              &=& 0 \\
              \\
H(X) &=& p_{Land} &* H(X|Land)) &+ p_{Tree} &* H(X|Tree) \\
             &=& \frac{5}{8}&*0.721 &+ \frac{3}{8}&*0 \\
             &=& 0.451205
\end{eqnarray}$$
<br>
<b>Information Gain</b> if split on Location = 1 - 0.451205 = 0.548795

              = -(\frac{1}{2}*log_2(\frac{1}{2}) + \frac{1}{2}*log_2(\frac{1}{2})) = 1$$

## <font color=' #7cb342 '>Pruning</font>

If a decision tree is allowed to grow as deeply as possible, then we run into the danger of overfitting. Hence, there are a few common stopping criteria that people use to prevent overfitting and make the tree generalizable:

* Maximum tree depth
* Minimum number of samples at one node
* Stop growing when the accuracy on the validation dataset does not improve





##<font color=' #7cb342 '> Pros</font>


Why use Decision Tree?

1.   **If you need to explain why your model works?**  Decision Tree is a white box type of ML algorithm. It shares internal decision-making logic, which is not available in the black box type of algorithms (most algorithms) such as Neural Network. 

2.   Its training time is typically faster. The time complexity of decision trees is a function of the number of records and number of attributes in the given data. 

##<font color=' #7cb342 '> Cons</font>

* Sensitive to noisy data. It can overfit noisy data.
* Might not be the most accurate model.

# <font color='blue'>Measurement Metrics</font> (When Accuracy is not Enough!)

In the health care field, it is not enough to know your model's accuracy.  Doctors need to know when you are accurate when your diagnosis is positive, and when you are accurate when your diagnosis is negative.
<br><br>
Why? Well sometimes you want to minimize the times when you tell your patients that they are sick when they are really healthy (**Type I error**, aka, false positive).  What's more important, you want to minimize the times when you tell your patients that they are healthy when they are already sick! (**Type II error**, aka, false negative)
<br><br>
Obviously we want to minimize both errors, but sometimes that is just impossible.  And sometimes it's just way more costly to one type of error and we want to minimize that first.
<br><br>
You will run into both errors if/when you take Statistics course.

## <font color=' #7cb342 '>Accuracy, Recall, and Precision</font>

accuracy: when your prediction is the same as the real nature
<br><br>
recall: your ability to find out all sick persons
<br><br>
precision: the correctness when you predict someone is sick

![alt text](https://i.ytimg.com/vi/RYFViaaJxE8/hqdefault.jpg)

#### Accuracy 
number of examples correctly predicted / total number of examples  
![alt text](https://wikimedia.org/api/rest_v1/media/math/render/svg/c72ec21ef2505c2d376e96197637fc64f75e5891)

```
metric = "ACC"
results = {}
results[metric] = (TP + TN) / (TP + TN + FP + FN)
print(f"{metric} is {results[metric]: .3f}")
```

#### Recall / Sensitivity / True Positive Rate
number of samples actually and predicted as  `Positive` / total number of samples actually `Positive`  
Also called **Sensitivity or Recall**.  
![alt text](https://wikimedia.org/api/rest_v1/media/math/render/svg/64d5540cbadeb83f864d7a731b7ab43cccd0f353)

```
# Sensitivity or Recall
metric = "TPR"
results[metric] = TP / (TP + FN)
print(f"{metric} is {results[metric]: .3f}")
```

#### True Negative Rate
number of samples actually and predicted as  `Negative` / total number of samples actually  `Negative`  
Also called **Specificity**.  
![alt text](https://wikimedia.org/api/rest_v1/media/math/render/svg/aac65b6586cb61c1268403895012c2a8f19b0c5e)


```
# Specificity
metric = "TNR"
results[metric] = TN / (TN + FP)
print(f"{metric} is {results[metric]: .3f}")
```

#### Precision / Positive Predictive Value
number of samples actually and predicted as  `Positive` / total number of samples predicted as `Positive`  
Also called **Precision**.  
![alt text](https://wikimedia.org/api/rest_v1/media/math/render/svg/699fcdb880b7f6a92742bc0845b8b60b59806a98)

```
# Precision
metric = "PPV"
results[metric] = TP / (TP + FP)
print(f"{metric} is {results[metric]: .3f}")
```

#### Negative Predictive Value
number of samples actually and predicted as  `Negative` / total number of samples predicted as `Negative`    
![alt text](https://wikimedia.org/api/rest_v1/media/math/render/svg/b6da0f59faffbc1704974e97847ab4fa599f5521)

```
metric = "NPV"
results[metric] = TN / (TN + FN)
print(f"{metric} is {results[metric]: .3f}")
```

#### F1 score
Harmonic Mean of Precision and Recall.  
![alt text](https://wikimedia.org/api/rest_v1/media/math/render/svg/5663ca95d471868169c4e4ea57c936f1b6f4a588)

```
metric = "F1"
results[metric] = 2 / (1 / results["PPV"] + 1 / results["TPR"])
print(f"{metric} is {results[metric]: .3f}")
```

```
print(f"Calculated and Actual Accuracy:                          {results['ACC']: .3f}, {metrics.accuracy_score(Y_test, Y_pred): .3f}")
print(f"Calculated and Actual Precision score:                   {results['PPV']: .3f}, {metrics.precision_score(Y_test, Y_pred): .3f}")
print(f"Calculated and Actual Recall score:                      {results['TPR']: .3f}, {metrics.recall_score(Y_test, Y_pred): .3f}")
print(f"Calculated and Actual F1 score:                          {results['F1']: .3f}, {metrics.f1_score(Y_test, Y_pred): .3f}")
```

## <font color=' #7cb342 '>ROC curve (Receiver Operating Characteristic curve)</font>
 
 Depending on the task at hand, accuracy may or may not be a good metric by which to judge your ML model by. In many problems, and particularly in medicine, incidences of positive and negative examples are not evenly distributed. 
 <br><br>
 For example, your data may contain <10% positive examples. If the algorithm classifies everything as negative, it will still achieve 90%+ accuracy. 
 <br><br>
 Generally, as the true positive rate increases, the false positive rate will also increase. The idea is to find an algorithm in which we can achieve a high TPR while minimizing the FPR. 
 <br><br>
 In many ML applications, and particularly in medicine the receiver operating characteristic (ROC) curve is widely used as an evaluation metric to avoid falling into this trap. Rather than measuring accuracy it plots the true positive rate (TPR = recall = TP/[TP+FN]) versus the false positive rate (FPR = FP/[FP + TN]).  In this way, the ROC curve allows us to visualize the true tradeoff between positive rate (TPR) and false positive rate (FPR).

#<font color='green'>Random Forest</font>

##<font color='green'>Ensemble Method</font>

A common criticism to decision trees is that once the training set is divided after answering a question, it is not possible to reconsider this decision. 
<br><br>
For example, if we divide men and women, every subsequent question would be only about men or women, and the method could not consider another type of question (say, age less than a year, irrespective of the gender). 
<br><br>
Random Forests try to introduce some level of randomization in each step, proposing alternative trees and combining them to get the final prediction. These types of algorithms that consider several classifiers answering the same question are called **ensemble methods**.

# <font color='orange'>Tree Regressor Intuition</font>

##When to use classification vs regression tree?


If the target variable is categorial, use classification tree/classifer.

If the target variable is numeric or continuous, use regression tree/regressor.


##Regression Tree Intuition (not required)

Constructing a decision tree for classification, we use Information Gain as splitting criterion.  The result obtained in the leaf node is calculated using "mode" of the observation.
<br><br>
Constructing a decision tree for regression, we use Standard Deviation Reduction (SDR) as splitting criterion.  Standard Deviation Reduction is the difference between the standard deviation of the target variable before the split and the standard deviation of the target variable after the split. The result obtained in the leaf node is computed using "mean/average" of the observation.
<br><br>
SDR: Standard Deviation Reduction
<br>
S(T): Standard Deviation before splitting
<br>
S(T, X): Standard Deviation after splitting based on feature X
<br>
