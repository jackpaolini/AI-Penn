# How Computer Works

Computer stores and processes data value.  Specifically data is stored in a hardware component, called <b>Disk</b> (also called <b>Hard Drive</b>).  

Before computer starts processing data, it loads data from Disk to another hardware component, called <b>Memory</b>.  Why?  Because Memory is faster than Disk.

When computer starts calculation, it loads data from Memory to another hardware component, called <b>Central Processing Unit (CPU)</b>.  Obviously this is a very important component!

When calculation is done, data is saved in the Memory, in case it's needed again soon.  Data will be saved on the Disk so that it won't be lost after shutdown.

### Built-in Functions

***capitalize()*** changes the first letter to capital letter.

```
sentence = "i am learning python crash course"
sentence = sentence.capitalize()
print(sentence)
```

**lower( )** converts any capital letter to small letter.



```
print(sentence.lower())
```

**upper( )** converts any small letter to capital letter.

```
print(sentence.upper())
```

**replace( )** function replaces the element with another element.

```
print(sentence.replace('crash course','jupyter notebook'))
```

**join( )** function can also be used to convert a list into a string.

```
a = ["I", "am", "learning", "python", "crash", "course"]
b = ' '.join(a)
print(b)
```

**split( )** function is used to convert a string back to a list. Think of it as the opposite of the **join()** function.

```
d = b.split(' ')
print(d)
```

In **split( )** function one can also specify the number of times you want to split the string or the number of elements the new returned list should conatin. The number of elements is always one more than the specified number this is because it is split the number of times specified.

```
e = b.split(' ', 2)
print(e)
```

**find( )** function returns the index value of the given data that is to found in the string. If it is not found it returns **-1**. Remember to not confuse the returned -1 for reverse indexing value.

```
print(sentence)
sentence.find('earn')
```
